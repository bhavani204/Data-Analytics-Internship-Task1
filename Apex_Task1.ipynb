```python
import pandas as pd
df = pd.read_excel("Sales_Dataset_With_Missing_And_Duplicates.xlsx")
df.info()
df.head(20)
```

```python
df.describe(include='all')
```

```python
print(df.isnull().sum())
```

```python

df['Customer Name']=df['Customer Name'].fillna(df['Customer Name'].mode()[0])
df['Product']=df['Product'].fillna(df['Product'].mode()[0])
df['Sales']=df['Sales'].fillna(df['Sales'].mean())
df['Profit']=df['Profit'].fillna(df['Profit'].median())
df['Order Date']=df['Order Date'].fillna(df['Order Date'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
Q1=df['Sales'].quantile(0.25)
Q3=df['Sales'].quantile(0.75)
IQR=Q3-Q1
lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR
print(lower_fence)
print(upper_fence)
```

```python
filter_data=df[df['Sales'].between(lower_fence,upper_fence)]
print(filter_data)
```

```python
df.to_excel("Sales_Dataset_With_Missing_And_Duplicates.xlsx", index=False)
```
